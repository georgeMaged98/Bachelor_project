# Predicting joint angle of  sensors for 1 subject

The goal of this notebook is to try and predict joint angles using the readings of electrodes where each 10 readings will be averaged for all electrodes.

**Helper Functions and basic imports**

```python
# Imports
%matplotlib inline
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score, auc, roc_curve
```

```python
def load_data_DB1(subject,exercise):
    data = loadmat(f'../Dataset/DB1/s{subject}/S{subject}_A1_E{exercise}.mat')
    return data
```

```python
def get_start_indexes(data,movement,target):
    start_indexes=[]
    for idx,move in enumerate(data[target]):
        if move[0] == movement and data[target][idx-1][0] == 0:
            start_indexes.append(idx)
        if len(start_indexes) == 10:
            break
    return start_indexes
```

```python
def avg_10_rows(data, n_cols):
    '''
    This function averages the values of 10 rows into 1 row for all columns. 
    It handles if the number of rows is not divisible by 10.
    It returns a dataframe where each row is the avg of 10 rows.
    '''
    
    # prepare the shape of input and output data
    n_rows = data.shape[0]
    n_tens = int(n_rows / 10)
    rem_length = n_rows % 10
    
    data_avg = np.zeros(shape=(n_tens+1, n_cols))
    
    # Avg every 10 rows and add result row to data_avg
    for i in np.arange(0, n_tens):
        start_idx = i * 10
        end_idx = start_idx + 20
        avg_row = data[start_idx:end_idx].mean(axis = 0)
        data_avg[i] = avg_row
    
    # Add average of remaining rows 
    avg_rem_rows = data[-rem_length:].mean(axis = 0)
    data_avg[-1] = avg_rem_rows
    return data_avg
```

```python
def train_model_glove(emg_data, glove_data):
    sensors_training_acc = []
    sensors_test_acc = []

    # Train a Random Forest regressor model for each of the 22 sensors
    for i in np.arange(0, 22):
        X = emg_data
        y = glove_data[:,i]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

        #Initiate the model
        rfr = RandomForestRegressor(n_estimators=100)
        # Fit the model 
        rfr.fit(X_train, y_train)

        training_score = round(rfr.score(X_train, y_train), 3)
        test_score = round(rfr.score(X_test, y_test), 3)

#         print(f"Training Accuracy score for sensor {i+1} is {training_score}")
#         print(f"Test Accuracy score for sensor {i+1} is {test_score}")

        sensors_training_acc.append(training_score)
        sensors_test_acc.append(test_score)
    
    acc_dict = {}
    acc_dict['training'] = sensors_training_acc
    acc_dict['test'] = sensors_test_acc
    return acc_dict
```

```python
def get_movement_df(data,movement,target):
    '''
    Returns the repetitions of a movement based on fixed length of repetition.
    if target is stimulus then each rep takes 5 seconds but if target is restimulus then each rep takes 2 seconds
    '''
    start_indexes = get_start_indexes(data,movement,target)
    df= pd.DataFrame()
    step = 500 if target == 'stimulus' else 200
    for i in start_indexes:
        new_row = pd.DataFrame(data['emg'][i:i+step])
        glove_data = pd.DataFrame(data['glove'][i:i+step])
        for j in np.arange(10,32):
            new_row[j] = glove_data[j-10]
        df = df.append(new_row,ignore_index=True)
#     target_movement = np.full((2000,1),movement,dtype=int)
#     df['target'] = target_movement
    return df
```

```python
def get_full_repetitions(data, movement):
    '''
    Returns a dataframe with the 10 repetitions of a movement  based on restimulus.
    The returned dataframe considers the varying length of each repetition.
    '''
    emg_df = pd.DataFrame()
    glove_df = pd.DataFrame()
    restimulus_indices = get_start_indexes(data=data, movement=movement, target="restimulus")
    for i in restimulus_indices:
        j = i
        while data["restimulus"][j] == movement:
            j+=1
        repetition_emg_df = pd.DataFrame(data["emg"][i:j])
        repetition_glove_df = pd.DataFrame(data["glove"][i:j])

        emg_df = emg_df.append(repetition_emg_df, ignore_index=True)
        glove_df = glove_df.append(repetition_glove_df, ignore_index=True)
        
#     col_names = list(range(0,df.shape[1] - 1))
#     col_names.append("target")
#     df = df.reindex(columns=col_names)
    return {
        "emg_df":emg_df,
        "glove_df":glove_df
    }
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

ex1_movements = np.arange(1,13)
ex2_movements = np.arange(1,18)
ex3_movements = np.arange(1,24)

models_dict_all_avg10 = {}

for label, model in models.items():
    models_dict_all_avg10[label] = np.zeros(shape=(22,27))

for sub in np.arange(1,28):
    print(f'Subject {sub}')

    emg_df = pd.DataFrame()
    glove_df = pd.DataFrame()
    
    
    data = load_data_DB1(subject=sub, exercise=1)
    
    for m in ex1_movements:
        move_df = get_full_repetitions(data=data, movement=m)
        emg_df = emg_df.append(move_df['emg_df'], ignore_index=True)
        glove_df = glove_df.append(move_df['glove_df'], ignore_index=True)    
        
    
    data = load_data_DB1(subject=sub, exercise=2)

    for m in ex2_movements:
        move_df = get_full_repetitions(data=data, movement=m)
        emg_df = emg_df.append(move_df['emg_df'], ignore_index=True)
        glove_df = glove_df.append(move_df['glove_df'], ignore_index=True)    
    
    
    data = load_data_DB1(subject=sub, exercise=3)
    
    for m in ex3_movements:
        move_df = get_full_repetitions(data=data, movement=m)
        emg_df = emg_df.append(move_df['emg_df'], ignore_index=True)
        glove_df = glove_df.append(move_df['glove_df'], ignore_index=True)    

    X = emg_df

    for i in np.arange(0, 22):

        y = glove_df.iloc[:, i]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

        for label, model in models.items():
            model.fit(X_train, y_train)
            y_preds = model.predict(X_test)
            
            # Average Every 10 rows
            n_rows10 = math.floor(len(y_test) / 10) + 1
            y_test10 = np.zeros(shape=(n_rows10,1))
            y_preds10 = np.zeros(shape=(n_rows10,1))

            for i in np.arange(0, n_rows10 -1):
                start = i*10
                end = start + 10
                y_test10[i] = np.mean(y_test[start:end])
                y_preds10[i] = np.mean(y_preds[start:end])

            rem_start = (n_rows10 - 1) * 10
            y_test10[-1] = np.mean(y_test[rem_start:])
            y_preds10[-1] = np.mean(y_preds[rem_start:])
            score = r2_score(y_test10, y_preds10)
            
            models_dict_all_avg10[label][i, sub-1] = score
    print(f"R^2 scores for subject {sub} for all sensors averaging every 10 rows are {models_dict_all_avg10['RandomForestRegressor'][:,sub-1]}")
```
